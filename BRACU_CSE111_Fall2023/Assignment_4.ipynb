```python
#task 1


class Student:
    def __init__(self,name,cg, credit = 9, dept = "CSE"):
        self.name = name
        self.cgpa = cg
        self.credit = credit
        self.department = dept
        self.scl = ""


    def checkScholarshipEligibility(self):
            if self.cgpa < 3.5 or self.credit < 10 :
                print(f"{self.name} is not eligible for scholarship.")
                self.scl = "No scholarship"
            elif self.cgpa>=3.5 and self.credit>10:
                if self.cgpa >=3.7:
                    print(f"{self.name} is eligible for Merit based scholarship.")
                    self.scl = "Merit based scholarship."
                elif  3.5 <= self.cgpa <3.7 :
                    print(f"{self.name} is eligible for Need-based scholarship.")
                    self.scl = "Need-based scholarship."

    def showDetails(self):
        print(f"Name: {self.name}")
        print(f"Department: {self.department}")
        print(f"CGPA: {self.cgpa}")
        print(f"Number of Credits: {self.credit}")
        print(f"Scholarship Status: {self.scl}")




print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()





```

```python
#task2


class Foodie:
    menu = {'Chicken Lollipop': 15, 'Beef Nugget': 20, 'Americano': 180, 'Red Velvet': 150, 'Prawn Tempura': 80, 'Saute Veg': 200}

    def __init__(self, name):
        self.name = name
        self.cart = {}
        self.tips = 0

    def show_orders(self):
        total_items = sum(self.cart.values())
        items_list = [item for item, qty in self.cart.items() for _ in range(qty)]
        total_spent = sum(self.menu[item] * self.cart[item] for item in self.cart) + self.tips

        print(f"{self.name} has {total_items} item(s) in the cart.")
        print(f"Items: {items_list}")
        print(f"Total spent: {total_spent}.")
        return ''

    def order(self, *args):
        for a in args:
            item, quantity = a.split('-')
            item = item.strip()
            quantity = int(quantity)
            if item in self.menu:
                if item in self.cart:
                    self.cart[item] += quantity
                else:
                    self.cart[item] = quantity
                print(f"Ordered - {item}, quantity - {quantity}, price (per Unit) - {self.menu[item]}.")
                print(f"Total price - {self.menu[item] * quantity}")
            else:
                print(f"{item} is not available in the menu.")

    def pay_tips(self, amount=0):
        if amount > 0:
            self.tips += amount
            print(f'Gives {amount}/- tips to the waiter.')
        else:
            print('No tips to the waiter.')

menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}


f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())


```

```python
#task 3
class Department:

    def __init__(self, name='ChE Department', sections=5):
        self.department = name
        self.sections = sections
        self.average = 0
        print(f"The {self.department} has {self.sections} sections.")

    def add_students(self, *students):
        self.students = students
        if len(self.students) == self.sections:
            totalStudents = 0
            for i in students:
                totalStudents += i
            self.average = round(totalStudents / self.sections, 2)
            print(f"The {self.department} Department has an average of {self.average} students in each section.")
        else:
            print(f"The {self.department} Department doesn't have {len(students)} sections.")

    def merge_Department(self, *departments):
        allStudents = self.average * self.sections
        for i in departments:
            print(f"{i.department} Department is merged to {self.department}.")
            allStudents += i.average * i.sections
        return f"Now the {self.department} has an average of {allStudents / self.sections} students in each section."

d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21, 30, 40, 36, 10, 32, 27, 51, 45, 15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
#
#task4

class Shopidify:
    def __init__(self, username = "Guest"):
        self.username = username
        self.cart = {}
        self.purchase_history = []
        if username == "Guest":
          print("Welcome to Shopidify")
        else:
          print(f"Welcome {self.username} to Shopidify")

    def add_to_cart(self, items, quantity = 1):
      if type(items) == str:
        if items not in self.cart:
          self.cart[items] = quantity
        else:
          self.cart[items] += quantity

      elif type(items) == list:
        for i in range(0, len(items)-1, 2):
          self.cart[items[i]] = items[i+1]

    def display_cart(self):
        print(f'Items in the cart for {self.username}:')

        for product, quantity in self.cart.items():
            print(f'- {product}: {quantity}x')


    def checkout(self):
        print(f"Checkout completed for {self.username} ")

    def display_history(self):
        print(f"Purchase history for {self.username}")
        print("Transection 1:")

        for product, quantity in self.cart.items():
            print(f'- {product}: {quantity}x')


guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3, "Goku Action Figure", 2, "Dumbbells-5kg", 2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
#task5
class Author:
    def __init__(self, name=None):
        self.name = name
        self.books = {}
        self.book_count = 0

    def setName(self, name):
        self.name = name

    def addBook(self, title, genre):
        if self.name is None:
            print("A book can not be added without author name")
        else:
            if genre not in self.books:
                self.books[genre] = [title]
            else:
                self.books[genre].append(title)
            self.book_count += 1

    def printDetail(self):
        print(f"Number of Book(s): {self.book_count}")
        print(f"Author Name: {self.name}")
        for genre, titles in self.books.items():
            print(f"{genre}: {', '.join(titles)}")

a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
