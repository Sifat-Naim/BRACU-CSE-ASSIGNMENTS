```python
#task1
class CellPackage:
    def __init__(self, price = 0, data = "",Talktime = 0,msg = 0,pr = "", Validity = 0 ):
        self.price = price
        self.talktime = Talktime
        self.sms = msg
        self.validity = Validity
        self.data = int(data.strip(" GB")) * 1024
        self.pr = int(pr[:-1])
        self.cashback = self.price * (self.pr/100)



    def printAttribute(self):
        if self.data > 0:
            print(f'Data = {self.data} MB')

        if self.talktime > 0:
            print(f'Talktime = {self.talktime} Minutes')

        if self.sms > 0:
            print(f'SMS/MMS = {self.sms}')

        if self.validity > 0:
            print(f'Validity = {self.validity} Days')

        if self.price > 0:
            print(f'--> Price = {self.price}')


        if self.cashback > 0:
            print(f'Buy now to get {int(self.cashback)} tk cashback.')

# Driver Code

pkg = CellPackage(150, '6 GB', 99, 20, '7%', 7)
print('============= Package 1 =============')
pkg.printAttribute()

pkg2 = CellPackage(700, '35 GB', 700, 0, '10%', 30)
print('============= Package 2 =============')
pkg2.printAttribute()

pkg4 = CellPackage(120, '0 GB', 190, 0, '0%', 10)
print('============= Package 3 =============')
pkg4.printAttribute()
```

```python
#task2

class box:
    def __init__(self,h):
        self.height = h[0]
        self.width = h[1]
        self.breadth = h[2]
        print("Creating a Box!")



    def boxDescription(self):
        print("Height:", self.height)
        print("Width:", self.width)
        print("Breadth:", self.breadth)

    def volume(self):

        return f"Volume of the box is {self.height * self.breadth * self.width} cubic units."




print("Box 1")
b1 = box([10,10,10])
print("=========================")
b1.boxDescription()
print(b1.volume())
print("-------------------------")
print("Box 2")
b2 = box((30,10,10))
print("=========================")
b2.boxDescription()
print(b2.volume())
b2.height = 300
print("Updating Box 2!")
print("Height:", b2.height)
print("Width:", b2.width)
print("Breadth:", b2.breadth)
volume = b2.height * b2.width * b2.breadth
print(f"Volume of the box is {volume} cubic units.")
print("-------------------------")
print("Box 3")
b3 = b2
b3.boxDescription()
print(b3.volume())
```

```python
#task4

class abcTech:
    def __init__(self, name, designation, department):
        self.name = name
        self.designation = designation
        self.department = department
        self.skills = []
        self.frameworks = []
        print(f"Welcome to abcTech, {name}")


    def addProgrammingSkills(self, skills):
        self.skills += skills


    def addFrameworks(self, frameworks):
        self.frameworks += frameworks

    def printInfo(self):
        print(f"Name:  {self.name}")
        print(f"Designation: {self.designation}")
        print(f"Department: {self.department}")
        print(f"Programming Skills: {', '.join(self.skills)}")
        print(f"Frameworks: {', '.join(self.frameworks)}")

    def calculateSalary(self, baseSalary, hours):
        self.salary = baseSalary
        self.hours = hours
        if hours > 144:
            self.salary = (((self.hours-144)*800) + self.salary)
        return self.salary

print("-------------------------")
b1 =abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java", "Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 =abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart", "Swift"])
b2.addFrameworks(["Flutter", "React Native"])
b2.addFrameworks(["Xamarin"])
b2.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b2.calculateSalary(103000, 123)}")
print("-------------------------")
```

```python
#task5

class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.grades = {}

    def calculateGPA(self, courses, semester):
        dict1 = {}
        sub_list = []
        total_gp = 0
        for i in courses:
            course, grade = i.split(": ")
            sub_list.append(course)
            total_gp += float(grade)*3
        grd = round(total_gp/ (len(sub_list)*3), 2)

        dict1[tuple(sub_list)] = grd
        self.grades[semester] = dict1


    def printDetails(self):

        print(f"Name: {self.name}")
        print(f"ID {self.id}")
        for k,v in self.grades.items():
            for x,y in v.items():
                print(f"Courses taken in {k}:")
                for i in x:
                    print(i)
                print(f"GPA {y}" )





# Write your code here
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()

```

```python
#task6

class Store:
    def __init__(self, initial_balance):
        print("New branch created!")
        self.balance = initial_balance
        self.total_items = 0
        self.inventory = {}

    def viewAllItems(self):
        if not self.inventory:
            print("There are no items in your inventory")
        else:
            print("All Items:", ', '.join(self.inventory.keys()))

    def viewAllItemDetails(self):
        if not self.inventory:
            print({})
        else:
            print(self.inventory)

    def add_item(self, item_info):
        item_name, stock, buying_price, selling_price = item_info
        if item_name in self.inventory:
            self.inventory[item_name]['stock'] += stock
        else:
            self.inventory[item_name] = {'stock': stock, 'buying_price': buying_price, 'selling_price': selling_price}
        self.balance -= stock * buying_price
        print(f"Item added: {item_name}")
        print(f"Current Balance: {self.balance}")

    def sell_item(self, item_name, quantity):
        if item_name in self.inventory and self.inventory[item_name]['stock'] >= quantity:
            selling_price = self.inventory[item_name]['selling_price']
            self.inventory[item_name]['stock'] -= quantity
            self.balance += quantity * selling_price
            print(f"Sold {quantity} {item_name} for {quantity * selling_price} - Current Balance: {self.balance}")
        else:
            print(f"Sorry! {item_name} is not available at your desired quantity. Currently we have: {self.inventory.get(item_name, {'stock': 0})['stock']}")
            print(f"Current Balance: {self.balance}")

    def restock_item(self, item_name, quantity):
        if item_name in self.inventory:
            buying_price = self.inventory[item_name]['buying_price']
            self.inventory[item_name]['stock'] += quantity
            self.balance -= quantity * buying_price
            print(f"Restocked item: {item_name}, Current Stock: {self.inventory[item_name]['stock']}")
            print(f"Current Balance: {self.balance}")

print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```
