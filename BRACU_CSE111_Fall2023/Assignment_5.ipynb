```python
#task2
class Student:
    def __init__(self,name,id,cgpa):
        self.name = name
        self.__id = id
        self.__cgpa = cgpa

    def setId(self,setid):
        self.__id = setid

    def get_id(self):
        return self.__id

    def get_cgpa(self):
        return self.__cgpa


class Department:
    def __init__(self,dept):
        self.dept = dept
        self.obj = []
        self.cse_id =[]



    def findStudent(self,user_id):
        for i in self.obj:
            if user_id in self.cse_id:
                print(f"Student info:")
                print(f"Student Name: {i.name}")
                print(f"ID: {i.get_id()}")
                print(f"CGPA:  {i.get_cgpa()}")
                return

        print("Student with this ID doesn't exist, Please give a valid ID")

    def addStudent(self,*obj):

       for i in obj:
          if i.get_id() in self.cse_id:
            print("Student with the same ID already exists, Please try with another ID")
            return
          self.obj.append(i)
          if i.get_id() not in self.cse_id:
            self.cse_id.append(i.get_id())

          print(f"Welcome to CSE department, {i.name}")


    def details(self):
        print(f"Department Name: {self.dept}")

        print(f"Number of student:{len(self.obj)}")
        print(f"Details of the students:")
        for i in self.obj:

            print(f"Student name: {i.name}, ID: {i.get_id()}, cgpa: {i.get_cgpa()}")





s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()

```

```python
#task 3
class Spaceship:
    def __init__(self,name,capacity):
        self.name = name
        self.capacity =capacity
        self.cargo = []
        self.w = 0

    def load_cargo(self,obj):
        self.obj = obj
        self.cargo.append(obj)
        if (self.capacity - self.w)< obj.getweight() :
            print("Warning: Unable to load Neutronium inside Falcon. ")
            print(f"Exceeds capacity by {obj.getweight()-(self.capacity - self.w)}.")


    def display_details(self):
        print(f"Spaceship Name: {self.name}")
        print(f"Capacity: {self.capacity}")

        name = []
        for i in self.cargo:
            self.w+= i.getweight()
            name.append(i.getname())
        print(f"Current Cargo Weight: {self.w}")
        print(f"Cargo: {name}")


class Cargo:
    def __init__(self,name,weight):
        self.__name = name
        self.__weight = weight

    def getname(self):
        return self.__name

    def getweight(self):
        return self.__weight






# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#task5

class Train:
    def __init__(self,*details):
        self.details = details
        self.tname = self.details[0]
        self.start = self.details[1]
        self.end = self.details[-1]

        print(f"Welcome aboard on {self.tname}")
        print(f"Start: {self.start}")
        print(f"Destination: {self.end}")
        self.passenger = []
    def addPassenger(self,*obj):
        for i in obj:
            if i.start==None:
                i.start=self.details[1]
            if i.destination==None:
                i.destination=self.details[-1]

            print(f"{i.name} welcome aboard")
            self.passenger.append(i)
    def allPassengerDetails(self):
        for i in self.passenger:
            fare = (self.details.index(i.destination) - self.details.index(i.start))*100

            print(f"Name: {i.name},Start: {i.start},Destination: {i.destination},Fair: ${fare}")





class Passenger:
    def __init__(self, name, start=None, destination=None):
        self.name = name
        self.start = start
        self.destination = destination


t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
#task6
class BracuStudent:
    def __init__(self, name, home):
        self.name = name
        self.home = home
        self.newpass = False

    def show_details(self):
        print(f"Student Name: {self.name}")
        print(f"Lives in {self.home}")
        print(f"Have Bus Pass? {self.newpass}")

    def get_pass(self):
        self.newpass = True

class BracuBus:
    def __init__(self, route, volume = 2):
        self.route = route
        self.volume = volume
        self.lst = []

    def show_details(self):
        print(f'''Bus Route: {self.route}
Passengers Count: {len(self.lst)} (Max: {self.volume})
Passengers On Board: {self.lst}
''')

    def board(self, *objects):
        if len(objects) == 0:
            print("No passengers!")
        else:
            for i in objects:
                if len(self.lst)<self.volume:
                    if i.home == self.route and i.newpass:
                        print(f"{i.name} boarded the bus.")
                        self.lst.append(i.name)
                    elif not i.newpass:
                        print("You don't have a bus pass!")
                    else:
                        print("You're on the wrong bus")
                else:
                    print("Bus is full!")





st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```
